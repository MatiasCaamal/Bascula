```python
import os
import shutil
import random
from pathlib import Path

# Ruta a tu carpeta original
origen = Path("DatasetsAumentados")
# Nueva carpeta para el dataset de clasificación
destino = Path("Frutas_cls")
porcentaje_val = 0.2  # 20% validación

# Crear estructura
for tipo in ["train", "val"]:
    for clase in os.listdir(origen):
        os.makedirs(destino / tipo / clase, exist_ok=True)

# Copiar imágenes a train y val
for clase in os.listdir(origen):
    ruta_clase = origen / clase
    imagenes = os.listdir(ruta_clase)
    random.shuffle(imagenes)

    split_idx = int(len(imagenes) * (1 - porcentaje_val))
    train_imgs = imagenes[:split_idx]
    val_imgs = imagenes[split_idx:]

    for img in train_imgs:
        shutil.copy(ruta_clase / img, destino / "train" / clase / img)
    for img in val_imgs:
        shutil.copy(ruta_clase / img, destino / "val" / clase / img)

print("Dataset preparado en formato YOLOv8 clasificación.")
```
