```python
!pip install ultralytics "numpy<2.0"

# ---IMPORTAMOS ---
import os
import shutil
import glob
import numpy as np
from pathlib import Path
from ultralytics import YOLO

# Rutas
ruta_origen = '/kaggle/input/frutas-dataset-completo/Frutas_cls'
ruta_trabajo = '/kaggle/working/Frutas_cls'

# Limpieza
if os.path.exists(ruta_trabajo):
    shutil.rmtree(ruta_trabajo)

for root, dirs, files in os.walk(ruta_origen):
    rel_path = os.path.relpath(root, ruta_origen)
    dest_dir = os.path.join(ruta_trabajo, rel_path)
    os.makedirs(dest_dir, exist_ok=True)

    for file in files:
        if file.endswith('.cache'): continue
        try:
            os.symlink(os.path.join(root, file), os.path.join(dest_dir, file))
        except: pass

print("Datos listos.")

# --- ENTRENAMIENTO---
print("Iniciando entrenamiento RÁPIDO...")
model = YOLO('yolov8n-cls.pt')

model.train(
    data=ruta_trabajo,
    epochs=30,
    imgsz=224,
    batch=128,
    project='/kaggle/working/resultados',
    name='Modelo_Final_Rapido',
    amp=False,
    workers=4
)
```
